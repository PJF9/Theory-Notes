## What is a Generative Problem?

So far we have seen `regression` and `classification` problem. What if we want out models to generate a new image (based on a Dataset), or in general to generate new Data. Those problems are called `generative` problems and to solve them we use GANs (**Generative Adversarial Networks**).

-

**More formally**: `Generative modeling` is an unsupervised learning task in machine learning that involves automatically discovering and learning the patterns in input data in such a way that the model can be used to generate new examples that plausibly could have been drawn from the original dataset. In fact, a really good generative model may be able to generate new examples that are not just plausible, but indistinguishable from real examples from the problem domain.

-

`GANs` are a clever way of training a `generative model` by framing the problem as a supervised learning problem with two sub-models: 
1. the **generator** model that we train to generate new examples, and 
2. the **discriminator** model that tries to classify examples as either _real_ (from the domain) or _fake_ (generated). 

The two models are trained together in a zero-sum game, adversarial, until the discriminator model is fooled about half the time, meaning the generator model is generating plausible examples.

## What Are Generative Models?

To fully comprehend the answer to this question we first need to understand the differences of `supervised` and `unsupervised` learning:

* In **supervised** learning problems (such us prediction/regression) the Dataset that the model is trained on is composed by _samples_ (X) and _labels_ (y) and the model is trained by showing examples of inputs, having it predict outputs, and correcting itself to make the outputs more like the expected outputs. We can say that in `supervised learning`, the goal is to learn a mapping from inputs X to outputs y, given a labeled set of input-output pairs.

* In **unsupervised** learning problems (such us clustering) the Dataset only contains _samples_ (X) and a model is constructed by extracting or summarizing the patterns in the input data. There is no correction of the model, as the model is not predicting anything.

-

Another distinction that need to be addressed is `Discriminative` vs. `Generative` Modeling:

* **Discriminative** modeling its the traditionally name for _Classification_ and this is because the model must discriminate examples of input variables across classes and choose the class that a given example belongs.

* **Generative** modeling are all the _unsupervised_ learning models that summarize the distribution of input variables and may be used to create or generate new examples in the input distribution.

## What Are Generative Adversarial Networks?

**Generative Adversarial Networks**, or GANs, are a deep-learning-based generative model. More generally, GANs are a model architecture for training a generative model, and the most common use are deep learning models for this architecture.

-

The GAN model architecture involves two sub-models: a `generator` model for generating new examples and a `discriminator` model for classifying whether generated examples are real, from the domain, or fake, generated by the generator model (it's in some sense a game between the two models).

## The Generator Model

The `generator model` takes a fixed-length random vector as input and **generates** a sample in the domain. The vector is drawn from randomly from a **Gaussian distribution**, and the vector is used to _seed_ the generative process. After training, points in this multidimensional vector space will correspond to points in the problem domain, forming a compressed representation of the data distribution.

* This vector space is referred to as a `latent space`, or a vector space comprised of latent variables. Latent variables, or hidden variables, are those variables that are important for a domain but are not directly observable.

* We often refer to latent variables (or latent space), as a `projection` (or compression) of the data distribution. In the case of GANs, the generator model applies **meaning** to points in a chosen latent space, such that new points drawn from that latent space can be provided to the generator model as input and used to generate new and different output examples.

The `Generator Model` can be visualize as:

## The Discriminator Model

The `discriminator model` takes an example from the domain as input (real or generated) and predicts a binary class label of _real_ or _fake_ (generated). The real example comes from the **training dataset** and the generated examples are output by the **generator model**.

After the training process, the `discriminator model` is **discarded** as we are interested in the generator.

* We propose that one way to build good image representations is by training `Generative Adversarial Networks` (GANs), and later reusing parts of the generator and discriminator networks as feature extractors for supervised tasks (**transfer learning**).

The `Discriminator Model` can be visualize as:
<center><img src="https://machinelearningmastery.com/wp-content/uploads/2019/04/Example-of-the-GAN-Discriminator-Model.png"></center>

## GANs as a Two Player Game

The two models, the `generator` and `discriminator`, are trained together. The generator generates a batch of samples, and these, along with real examples from the domain, are provided to the discriminator and classified as _real_ or _fake_.

The `discriminator` is then updated to **get better at discriminating real and fake samples** in the next round, and the `generator` is updated based on **how well, or not, the generated samples fooled** the discriminator.

-

In this way, the two models are `competing` against each other, they are adversarial in the game theory sense, and are playing a zero-sum game.

In this case, `zero-sum` means that when the discriminator successfully identifies real and fake samples, no change is needed to the model parameters, whereas the generator is penalized with large updates to model parameters and vise versa.

-

Idealy we wanto the `discriminator` model to be totalu **unsure** (probability 50% for _real_ and _fake_).

## GANs and Convolutional Neural Networks

`GANs` typically work with **image data** and use Convolutional Neural Networks, or `CNNs`, as the generator and discriminator models.

The main reason is because of the remarkable progress using CNNs more generally to achieve `state-of-the-art` results on a suite of **computer vision** tasks such as object detection and face recognition.

## Conditional GANs

The `generative model` can be trained to generate new examples from the input domain, where the input, the random vector from the latent space, is provided with (conditioned by) some **additional input**.

The additional input could be a _class value_, such as male or female in the generation of photographs of people, or a _digit_, in the case of generating images of handwritten digits.

* GANs can be extended to a `conditional model` if both the `generator` and `discriminator` are conditioned on some extra information $y$. We can perform the conditioning by feeding $y$ into the both the `discriminator` and `generator` as an additional input layer.

-

Taken one step further, the `GAN` models can be **conditioned** on an example from the domain, such as an image. This allows for applications of `GANs` such as:
1. `text-to-image` translation
2. `image-to-image` translation.
3. `style transfer`
4. `photo colorization`

In the case of `conditional GANs` for `image-to-image` translation, such as transforming day to night, the `discriminator` is provided examples of real and generated nighttime photos as well as (**conditioned** on) real daytime photos as input. The `generator` is provided with a random vector from the latent space as well as (**conditioned** on) real daytime photos as input.

## Data Augmentation as GAN

`Data augmentation` results in better performing models, both increasing model skill and providing a regularizing effect, reducing generalization error. It works by creating new, artificial but plausible examples from the input problem domain on which the model is trained.

The techniques are primitive in the case of image data, involving **crops**, **flips**, **zooms**, and other **simple transforms** of existing images in the training dataset.

Successful `generative modeling` provides an alternative and potentially more **domain-specific** approach for `data augmentation`. In fact, data augmentation is a simplified version of generative modeling, although it is rarely described this way.

## Discriminator Network

```python
# Create a Helping Function
def conv_block(in_channels, out_channels):
    return nn.Sequential(
        nn.Conv2d(in_channels=in_channels,
                  out_channels=out_channels,
                  kernel_size=(4, 4),
                  stride=2,
                  padding=1,
                  bias=False),
        nn.BatchNorm2d(out_channels),
        nn.LeakyReLU(negative_slope=0.2, inplace=True)
    )
```

```python
class DiscriminatorN(nn.Module):
    def __init__(self, hidden_units_1, hidden_units_2, hidden_units_3, hidden_units_4):

        super().__init__()

        # First Conv Block: Output: (`batch_size`, `hidden_units_1`, 32, 32)
        self.conv_block_1 = conv_block(3, hidden_units_1)

        # Second Conv Block: Output: (`batch_size`, `hidden_units_2`, 16, 16)
        self.conv_block_2 = conv_block(hidden_units_1, hidden_units_2)

        # Third Conv Block: Output: (`batch_size`, `hidden_units_3`, 8, 8)
        self.conv_block_3 = conv_block(hidden_units_2, hidden_units_3)

        # Fourth Conv Block: Output: (`batch_size`, `hidden_units_4`, 4, 4)
        self.conv_block_4 = conv_block(hidden_units_3, hidden_units_4)

        # Last Conv Block: Output: (`batch_size`, 1, 1, 1)
        self.conv_block_5 = nn.Conv2d(in_channels=hidden_units_4,
                                      out_channels=1,
                                      kernel_size=(4, 4),
                                      stride=1,
                                      padding=0,
                                      bias=False)

        # Classifier Block: Output: (`batch_size`, 1)
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.classifier(self.conv_block_5(self.conv_block_4(self.conv_block_3(self.conv_block_2(self.conv_block_1(x))))))
```

## Generator Network

The input to the `generator` is a latent tensor (which is used as a seed for generating an image). The `generator` will convert the **latent tensor** of shape (128, 1, 1) into an **image tensor** of shape (3, 64, 64). To achive this, we'll use the `ConvTranspose2d` layer from PyTorch, which is performs to as a **transposed convolution**.

Usfull animations for **deconvolution** can be found: https://github.com/vdumoulin/conv_arithmetic/blob/master/README.md#transposed-convolution-animations

-

The Model is going to be a reverse process of the `dicriminator`.

```python
# Creating a Helping Function
def deconv_block(in_channels, out_channels):
    return nn.Sequential(
        nn.ConvTranspose2d(in_channels=in_channels,
                           out_channels=out_channels,
                           kernel_size=(4, 4),
                           stride=2,
                           padding=1,
                           bias=False),
        nn.BatchNorm2d(out_channels),
        nn.ReLU(inplace=True)
    )
```

```python
# Generally the higher the `latent_size` the better
latent_size = 128
```

```python
class GeneratorN(nn.Module):
    def __init__(self, hidden_units_1, hidden_units_2, hidden_units_3, hidden_units_4):
        super().__init__()

        # First Deconv Block: From (`batch_szie`, `latent_size`, 1, 1) to (`batch_szie`, `hidden_units_1`, 4, 4)
        self.deconv_block_1 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=latent_size,
                               out_channels=hidden_units_1,
                               kernel_size=(4, 4),
                               stride=1,
                               padding=0,
                               bias=False),
            nn.BatchNorm2d(hidden_units_1),
            nn.ReLU(inplace=True)
        )

        # Second Deconv Block: Output: (`batch_size`, `hidden_units_2`, 8, 8)
        self.deconv_block_2 = deconv_block(hidden_units_1, hidden_units_2)

        # Third Deconv Block: Output: (`batch_size`, `hidden_units_3`, 16, 16)
        self.deconv_block_3 = deconv_block(hidden_units_2, hidden_units_3)

        # Fourth Deconv Block: Output: (`batch_size`, `hidden_units_4`, 32, 32)
        self.deconv_block_4 = deconv_block(hidden_units_3, hidden_units_4)

        # Last Deconv Block: Output: (`batch_size`, 3, 64, 64)
        self.deconv_block_5 = nn.ConvTranspose2d(in_channels=hidden_units_4,
                                                 out_channels=3,
                                                 kernel_size=(4, 4),
                                                 stride=2,
                                                 padding=1,
                                                 bias=False)

        # Non-Linear Activation Function
        self.non_linearity = nn.Tanh()

    def forward(self, x):
        return self.non_linearity(self.deconv_block_5(self.deconv_block_4(self.deconv_block_3(self.deconv_block_2(self.deconv_block_1(x))))))
```

The `ReLU` activation is used in the `generator` with the exception of the output layer which uses the `Tanh` function. We observed that using a bounded activation allowed the model to learn more quickly to saturate and cover the color space of the training distribution.

-

Note that since the outputs of the `TanH` activation lie in the range `[-1,1]`, we have applied the **similar transformation** to the images in the training dataset.

## Discriminator Training

Since the `discriminator` is a **binary classification** model, we can use the **binary cross entropy** loss function to quantify how well it is able to differentiate between real and generated images.

```python
def train_discriminator(discriminator_model,
                        generator_model,
                        real_image_batch,
                        opt_discr,
                        latent_size,
                        loss_fn):

    ### 1. Setting things up ###

    # Setting Model's Devices
    discr_device = next(discriminator_model.parameters()).device
    gen_device = next(generator_model.parameters()).device


    ### 2. Getting Discriminator's Loss on Real Images ###

    # `real_logits`: {1: if `real_image` classified correctly, 0: otherwise}
    real_logits = discriminator_model(real_image_batch.to(discr_device))
    real_labels = torch.ones(real_image_batch.shape[0], 1, device=discr_device) # shape: (`batch_size`, 1)

    real_loss = loss_fn(real_logits, real_labels)
    real_score = torch.mean(real_logits).item() # Average Score per Batch (As close to 1 as possible)


    ### 3. Getting Discriminator's Loss on Generated Images ###

    # Generate `fake` images
    latent = torch.randn(real_image_batch.shape[0], latent_size, 1, 1)
    generated_images = generator_model(latent.to(gen_device))

    # `fake_logits`: {0: if `fake_image` classified correctly, 1: otherwise}
    fake_logits = discriminator_model(generated_images.to(discr_device))
    fake_labels = torch.zeros(generated_images.shape[0], 1, device=discr_device) # shape: (`batch_size`, 1)

    fake_loss = loss_fn(fake_logits, fake_labels)
    fake_score = torch.mean(fake_logits).item() # (As close to 0 as possible)


    ### 4. Updating Discriminator's Weights ###

    loss = real_loss + fake_loss
    
    opt_discr.zero_grad()
    loss.backward()
    opt_discr.step()

    return round(loss.item(), 4), round(real_score, 4), round(fake_score, 4)
```

## Generator Training

Since the outputs of the `generator` are images, it's not obvious how we can train the generator. This is where we employ a rather elegant trick, which is to use the `discriminator` as a part of the loss function. Here's how it works:

1. **Generate** a batch of images using the `generator` and pass them into the `discriminator`.

2. **Calculate** the loss by setting the target labels to 1 (aka real), because the `generator`'s objective is to _fool_ the `discriminator`.

3. **Use** the loss to perform gradient descent on the `generator` weights, so it gets better at generating real-like images to _fool_ the `discriminator`.

```python
def train_generator(discriminator_model,
                    generator_model,
                    opt_disc,
                    opt_gen,
                    batch_size,
                    latent_size,
                    loss_fn):
    
    ### 1. Setting things up ###
    discr_device = next(discriminator_model.parameters()).device
    gen_device = next(generator_model.parameters()).device


    ### 2. Generating Fake Images ###
    latent = torch.randn(batch_size, latent_size, 1, 1)
    fake_images = generator_model(latent.to(gen_device))


    ### 3. Passing the Fake Images into the Dicriminator (Trying to Fool it) ###
    logits = discriminator_model(fake_images.to(discr_device))

    # `labels`: {1: if `fake_image` classified incorrectly, 0: otherwise}
    #   (because we want the generator to 'fool' the discriminator)
    labels = torch.ones(batch_size, 1, device=discr_device)

    loss = loss_fn(logits, labels)


    ### 4. Updating Discriminator's Weights ###
    opt_gen.zero_grad()
    opt_disc.zero_grad()
    loss.backward()
    opt_gen.step()

    return round(loss.item(), 4)
```

## Creating a Training Loop

Let's define a `fit` function to train the discriminator and generator for each batch of training data. We'll use the `Adam optimizer` with some **custom parameters (betas)** that are known to work well for GANs. We will also save some sample generated images at regular intervals for inspection.

```python
def fit(discriminator_model,
        generator_model,
        epochs, train_dl,
        loss_fn,
        opt_discr,
        opt_gen,
        latent_size,
        print_per_epoch=1,
        print_total_batches=None,
        start_i=0):

    # Setting some evaluation Lists
    losses_discr, losses_gen = [], []
    real_scores, fake_scores = [], []

    # Getting Batch size
    batch_size = train_dl.batch_size

    print("Starting Process...")

    for epoch in tqdm(range(1, epochs + 1)):
        print(f"Starting Epoch: {epoch}...")
        dummy = 0

        for batch_num, (real_image, _) in enumerate(train_dl, start=1):
            # Training the Discriminator
            loss_discr, real_score, fake_score = train_discriminator(discriminator_model,
                                                                           generator_model,
                                                                           real_image,
                                                                           opt_discr,
                                                                           latent_size,
                                                                           loss_fn)
            # Training the Generator
            loss_gen = train_generator(discriminator_model,
                                       generator_model,
                                       opt_discr,
                                       opt_gen,
                                       batch_size,
                                       latent_size,
                                       loss_fn)
            
            if print_total_batches and (batch_num % (len(train_dl) // print_total_batches) == 0):
                dummy += 1
                print(f"\t{dummy}) Looked at {batch_num}/{len(train_dl)} training batches...")
            
        
        # Updating the Lists once both Models are trained on the Dataset
        losses_discr.append(loss_discr)
        losses_gen.append(loss_gen)
        real_scores.append(real_score)
        fake_scores.append(fake_score)

        if dummy != 0:
            print('-' * 93)

        # Printing Results
        if epoch % print_per_epoch == 0:
            save_samples(generator, epoch+start_i, fixed_latent, fake_images_path, False)
            print('-' * 93)
            print(
                f"-> Epoch: {epoch} | "
                f"Discr_Loss: {loss_discr} | "
                f"Gen_Loss: {loss_gen} | "
                f"Real_Score: {real_score} | "
                f"Fake_Score: {fake_score}")
            print('-' * 93)
            print("\n")

    print("Process Completed Succesfully...")

    return losses_discr, losses_gen, real_scores, fake_scores
```

## Setting Loss Function and Optimizers

```python
loss_fn = nn.BCELoss()
opt_discr = optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))
opt_gen = optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
```

## Training the Models

```python
history = fit(discriminator, generator, 25, train_dl, loss_fn, opt_discr, opt_gen, latent_size, 1, 2)
```
