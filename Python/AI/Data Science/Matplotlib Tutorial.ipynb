## What is Matploylib
Matploylib is a low lever graph that serves as a visualization utility. Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility. One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.

### Basic Graph

```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot - For more informations.

import matplotlib.pyplot as plt                      # To plot our graph we need to import the .pyplot class from matplotlib.
import numpy as np

x_axis = np.array([1, 2, 3, 4, 5, 6])
y_axis = np.array([2, 4, 6, 8, 10, 12])

plt.figure(figsize=(5, 4), dpi=100)                  # Resizing the size of the plot. The first parameter is the ration
                                                     #  of the axis and then the pixels of the graph.


plt.plot(x_axis, y_axis, color='g', marker='.',      # Creating the plotting graph base of x_axis and y_axis (First two
    linewidth=2, label="A Simple Line", ls=':',      #   parameters are the values of each axis. Then the color of the
    markerfacecolor='k', markeredgecolor='k')        #   graph, then points, the width of the graph, the label of the
                                                     #   graph, the style, the marker inside color and the outside color.
                                                     #   The colors are: b-blue, g-green, r-red, c-cyan, m-magenda,
                                                     #   y-yellow, k-black, w-white, the styles are '-': solid line,
                                                     #   '--': dashed line, '-.':dashed dot line, ':': dotted line and
                                                     #   the markers are '.',',','o','v','^','<','>','1','2','3','4'
                                                     #   '*','+','D','|','_').

y2 = 10 * np.log(x_axis) - 2                         # Creating a second line at our graph.
plt.plot(x_axis, y2, color='b', marker='*',
    linewidth=2, label="A Simple Log Line", ls='--',
    markerfacecolor='k', markeredgecolor='k')

plt.title("A Simple Graph", loc="center", pad=9.0,   # Setting a title (loc is the location on the graph, options are
    fontdict={"fontsize": 15, "fontname":"Arial"})   #  'center': default, 'left', 'right' and the other parameter is
                                                     #   how away the title will be from the graph, default is 6.0).
                                                     #   Another parameter is the fontdict, which is a dictionary that
                                                     #   manipulates the text of the title.
plt.xlabel("X Axis", loc="left",                     # Creating the X axis.
    fontdict={"fontsize": 12})
plt.ylabel("Y Axis", loc="bottom",                   # Creating the Y axis, loc parameters are 'bottom', 'center' and 'top'.
    fontdict={"fontsize":12})
plt.xticks(np.array([0, 1, 2, 3, 4, 5, 6]),          # Modifying the x axis.
    labels=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
plt.yticks(np.array([0, 1, 2, 3, 4, 6, 8, 10, 12]))  # Modifying the y axis.

plt.legend(loc="upper left",      # Creating a legend for our graph.
    fontsize=11, labelcolor="gray", shadow=True,
    facecolor='white')

plt.savefig("first_graph.png", dpi=500)              # Saving the graph on the current directory. Can specify the pixels too.
plt.show()                                           # Plotting the graph.
```

## Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array(["Greece", "Germany", "France", "Portugal", "Italy"])
y_axis = np.array([70, 20, 40, 60, 78])

plt.bar(x_axis, y_axis, width=0.5, align='center',                      # Creating the bar chart. Must of parameters are the
    color=['b', 'y', 'w', 'r', 'g'], edgecolor='k',                     #  same as before, except hatch:{\|/-+xoO.*}.
    linewidth=0.9, ls='--', hatch='/')

plt.show()
```
